In this notebook, we will calculate the predicted number of deaths in 2020.

# **Preliminary steps**

```python
### We import the packages

# To manage dataframes
import numpy as np
import pandas as pd
import math
```

```python
### We import previous dataframes

%store -r pred_1_women_f1
%store -r pred_1_women_f1_lower
%store -r pred_1_women_f1_upper
%store -r pred_1_men_f1
%store -r pred_1_men_f1_lower
%store -r pred_1_men_f1_upper

%store -r pred_1_women_f2
%store -r pred_1_women_f2_lower
%store -r pred_1_women_f2_upper
%store -r pred_1_men_f2
%store -r pred_1_men_f2_lower
%store -r pred_1_men_f2_upper

%store -r pred_2_women_f1
%store -r pred_2_women_f1_lower
%store -r pred_2_women_f1_upper
%store -r pred_2_men_f1
%store -r pred_2_men_f1_lower
%store -r pred_2_men_f1_upper

%store -r pred_2_women_f2
%store -r pred_2_women_f2_lower
%store -r pred_2_women_f2_upper
%store -r pred_2_men_f2
%store -r pred_2_men_f2_lower
%store -r pred_2_men_f2_upper

%store -r population_2020
```

# **Function to create the table**

```python
def table(pred_ratios_women, pred_ratios_women_lower, pred_ratios_women_upper, pred_ratios_men, pred_ratios_men_lower, pred_ratios_men_upper):
    deaths = population_2020.iloc[:,:100].copy()
    deaths_lower = population_2020.iloc[:,:100].copy()
    deaths_upper = population_2020.iloc[:,:100].copy()
    for i in range(100) :
        deaths.iloc[2,i] = math.ceil(deaths.iloc[2,i] * pred_ratios_women.iloc[0,i] / 100000)
        deaths.iloc[1,i] = math.ceil(deaths.iloc[1,i] * pred_ratios_men.iloc[0,i] /100000)
        deaths.iloc[0,i] = deaths.iloc[1,i] + deaths.iloc[2,i]
        
        deaths_lower.iloc[2,i] = math.ceil(deaths_lower.iloc[2,i] * pred_ratios_women_lower.iloc[0,i] / 100000)
        deaths_lower.iloc[1,i] = math.ceil(deaths_lower.iloc[1,i] * pred_ratios_men_lower.iloc[0,i] /100000)
        deaths_lower.iloc[0,i] = deaths_lower.iloc[1,i] + deaths_lower.iloc[2,i]
        
        deaths_upper.iloc[2,i] = math.ceil(deaths_upper.iloc[2,i] * pred_ratios_women_upper.iloc[0,i] / 100000)
        deaths_upper.iloc[1,i] = math.ceil(deaths_upper.iloc[1,i] * pred_ratios_men_upper.iloc[0,i] /100000)
        deaths_upper.iloc[0,i] = deaths_upper.iloc[1,i] + deaths_upper.iloc[2,i]
        
    table = pd.DataFrame(index = pd.IntervalIndex.from_breaks([0, 64, 100]), columns = ["Population", "Predicted deaths", "Observed deaths", "Difference", "Difference (in %)"])
    Total = pd.DataFrame(index = ["Total"], columns = ["Population", "Predicted deaths", "Observed deaths", "Difference", "Difference (in %)"])
    
    table.iloc[0,0] = sum(population_2020.iloc[0,:65])
    table.iloc[0,1] = [sum(deaths_lower.iloc[0,:65]),sum(deaths_upper.iloc[0,:65])]
    table.iloc[0,2] = 92478
    table.iloc[0,3] = [92478-sum(deaths_upper.iloc[0,:65]),92478-sum(deaths_lower.iloc[0,:65])]
    table.iloc[0,4] = [str(round(100*(92478-sum(deaths_upper.iloc[0,:65]))/sum(deaths_upper.iloc[0,:65]),2))+" %",str(round(100*(92478-sum(deaths_lower.iloc[0,:65]))/sum(deaths_lower.iloc[0,:65]),2))+" %"]
    
    table.iloc[1,0] = sum(population_2020.iloc[0,65:100])
    table.iloc[1,1] = [sum(deaths_lower.iloc[0,65:100]),sum(deaths_upper.iloc[0,65:100])]
    table.iloc[1,2] = 561538
    table.iloc[1,3] = [561538-sum(deaths_upper.iloc[0,65:100]),561538-sum(deaths_lower.iloc[0,65:100])]
    table.iloc[1,4] = [str(round(100*(561538-sum(deaths_upper.iloc[0,65:100]))/sum(deaths_upper.iloc[0,65:100]),2))+" %",str(round(100*(561538-sum(deaths_lower.iloc[0,65:100]))/sum(deaths_lower.iloc[0,65:100]),2))+" %"]
    
    Total.iloc[0,0] = sum(population_2020.iloc[0,:100])
    Total.iloc[0,1] = [sum(deaths_lower.iloc[0,:100]),sum(deaths_upper.iloc[0,:100])]
    Total.iloc[0,2] = 92478+561538
    Total.iloc[0,3] = [92478+561538-sum(deaths_upper.iloc[0,:100]),92478+561538-sum(deaths_lower.iloc[0,:100])]
    Total.iloc[0,4] = [str(round(100*(92478+561538-sum(deaths_upper.iloc[0,:100]))/sum(deaths_upper.iloc[0,:100]),2))+" %",str(round(100*(92478+561538-sum(deaths_lower.iloc[0,:100]))/sum(deaths_lower.iloc[0,:100]),2))+" %"]
    
    table = pd.concat([table,Total], axis=0)
    return table
```

# **First database**

```python
# Results for f1 function

table(pred_1_women_f1, pred_1_women_f1_lower, pred_1_women_f1_upper, pred_1_men_f1, pred_1_men_f1_lower, pred_1_men_f1_upper)
```
