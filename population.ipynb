In this notebook, we will import and clean our databases that contain population numbers in France from 2017 to 2020.

# **Preliminary steps**

```python
### We first import our packages

# To manage dataframes
import numpy as np
import pandas as pd
```

# **Functions to clean dataframes**

```python
# Function to clean population databases
def cleaning_population(df, new_index, col_to_drop) :
    df.index = new_index
    df.drop(col_to_drop, axis = 1, inplace = True)
    return(df)
```

# **Data cleaning**

```python
# We import our databases

population_2017 = pd.read_csv("https://minio.lab.sspcloud.fr/juleschpn/Population_2017.csv", sep=';', encoding = 'latin-1', thousands = ' ',low_memory = False)
population_2018 = pd.read_csv("https://minio.lab.sspcloud.fr/juleschpn/Population_2018.csv", sep=';', encoding = 'latin-1', thousands = ' ',low_memory = False)
population_2019 = pd.read_csv("https://minio.lab.sspcloud.fr/juleschpn/Population_2019.csv", sep=';', encoding = 'latin-1', thousands = ' ',low_memory = False)
population_2020 = pd.read_csv("https://minio.lab.sspcloud.fr/juleschpn/Population_2020.csv", sep=';', encoding = 'latin-1', thousands = ' ',low_memory = False)
population_2020.head(n = 5)
```
